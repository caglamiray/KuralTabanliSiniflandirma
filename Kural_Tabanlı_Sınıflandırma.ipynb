```python
#Görev 2: COUNTRY, SOURCE, SEX, AGE kırılımında ortalama kazançlar nedir?

new_df = df.groupby([ "COUNTRY" , "SOURCE" , "SEX" , "AGE" ]).agg({ "PRICE" : "mean"})

new_df
```

```python
#Görev 3: Çıktıyı PRICE’a göre sıralayınız.
#• Önceki sorudaki çıktıyı daha iyi görebilmek için sort_values metodunu azalan olacak şekilde PRICE’a göre uygulayınız.
#• Çıktıyı agg_df olarak kaydediniz. 
```

```python
agg_df = new_df.sort_values("PRICE" , ascending = False)

agg_df
```

```python
len(agg_df.columns)  #çünkü orada sadece price değişkendir diğerleri ona işaret eder
```

```python
agg_df = agg_df.reset_index()  #şimdi hepsi değişken oldu

agg_df
```

```python
#Görev 5: Age değişkenini kategorik değişkene çeviriniz ve agg_df’e ekleyiniz.

#• Age sayısal değişkenini kategorik değişkene çeviriniz.
#• Aralıkları ikna edici şekilde oluşturunuz.
#• Örneğin: ‘0_18', ‘19_23', '24_30', '31_40', '41_7'
```

```python
#Age değişkeninin nerelerden bölüneceğini belirtelim:

my_bins = [0, 18, 23, 30, 40, agg_df["AGE"].max()]  #en büyük sayıyı istedik

my_bins
```

```python
#Bölünen noktalara göre isimlendirme yapalım:

mylabels = ["0_18" , "19_23" , "24_30" , "31_40" , "41_" + str(agg_df["AGE"].max())]

mylabels
```

```python
#age'i bölelim: 

pd.cut(agg_df["AGE"] , bins = my_bins , labels = mylabels).sort_values()
```

```python
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"] , bins = my_bins , labels = mylabels).sort_values()

agg_df
```

```python
#Görev 6: Yeni seviye tabanlı müşterileri (persona) tanımlayınız.

#• Yeni seviye tabanlı müşterileri (persona) tanımlayınız ve veri setine değişken olarak ekleyiniz.
#• Yeni eklenecek değişkenin adı: customers_level_based
#• Önceki soruda elde edeceğiniz çıktıdaki gözlemleri bir araya getirerek customers_level_based değişkenini 
#oluşturmanız gerekmektedir.
```

```python
#Dikkat! List comprehension ile customers_level_based değerleri oluşturulduktan sonra bu 
#değerlerin tekilleştirilmesi gerekmektedir. 
#Örneğin birden fazla şu ifadeden olabilir: USA_ANDROID_MALE_0_18. Bunları groupby'a 
#alıp price ortalamalarını almak gerekmektedir
```

```python
agg_cols = ["COUNTRY" , "SOURCE" , "SEX" , "AGE_CAT"]
```

```python
[col[0].upper() + "_" + col[1].upper() + "_" + col[2].upper() + "_" + col[3].upper() for col in agg_df[agg_cols].values]
```

```python
#Gereksiz değişkenleri çıkaralım

agg_df.head()
```

```python
agg_df = agg_df[ ["CUSTOMER_LEVEL_BASED" , "PRICE"] ]

agg_df
```

```python
agg_df.groupby("CUSTOMER_LEVEL_BASED")["PRICE"].mean()
```

```python
agg_df = agg_df.groupby("CUSTOMER_LEVEL_BASED")["PRICE"].mean().reset_index()

agg_df
```

```python
#Görev 7: Yeni müşterileri (personaları) segmentlere ayırınız.

#• Yeni müşterileri (Örnek: USA_ANDROID_MALE_0_18) PRICE’a göre 4 segmente ayırınız.
#• Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df’e ekleyiniz.
#• Segmentleri betimleyiniz (Segmentlere göre group by yapıp price mean, max, sum’larını alınız)
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df.PRICE, 4, labels = ["D" , "C" , "B" , "A"])

agg_df
```

```python
agg_df.groupby("SEGMENT").agg({"PRICE" : ["mean" , "sum" , "min" , "max"]})
```
